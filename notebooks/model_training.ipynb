```python
# Define which columns should be ordinal-encoded and which should be scaled
# List of numerical columns that will be scaled
numerical_cols = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5',
                    'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']


# List of categorical columns that will be ordinal-encoded
categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']


```

```python
# Import necessary tools from scikit-learn
from sklearn.preprocessing import StandardScaler   # StandardScaler is used for scaling (normalizing) numerical features.

# Import tools for constructing pipelines and column transformations
from sklearn.pipeline import Pipeline              # Pipeline is used to chain multiple transformers and an estimator into a single object.
from sklearn.compose import ColumnTransformer      # ColumnTransformer is used to apply different transformers to different columns of a DataFrame.
```

```python
# Define a pipeline for numerical feature scaling
num_pipeline = Pipeline(
    steps=[
        ('scaler', StandardScaler())  # StandardScaler is used to scale (normalize) numerical features.
    ]
)
  
```

```python
# For Categorigal Pipeline
cat_pipeline=Pipeline(
    steps=[
    ('scaler',StandardScaler())
    ]
)
```

```python
# Define a ColumnTransformer for preprocessing numerical and categorical features separately
preprocessor = ColumnTransformer([
    ('num_pipeline', num_pipeline, numerical_cols),  # Apply num_pipeline to numerical_cols
    ('cat_pipeline', cat_pipeline, categorical_cols)  # Apply cat_pipeline to categorical_cols
])
```

```python
# Import train_test_split function from scikit-learn
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# X_train: Features for training
# X_test: Features for testing
# y_train: Target variable for training
# y_test: Target variable for testing
# test_size=0.30: Splitting the data into 70% training and 30% testing
# random_state=30: Random seed for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=30)
```

```python
# Apply the preprocessor (ColumnTransformer) to transform the training features
preprocessor.fit_transform(X_train)
```

```python
# Apply the preprocessor (ColumnTransformer) to transform the testing features
preprocessor.transform(X_test)
```

```python
# Get the output feature names after applying the preprocessor (ColumnTransformer)
preprocessor.get_feature_names_out()
```

```python
# Apply preprocessor to transform and create DataFrames for training and testing features
X_train = pd.DataFrame(preprocessor.fit_transform(X_train), columns=preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test), columns=preprocessor.get_feature_names_out())
```

```python
# Import necessary libraries for model training and evaluation
from sklearn.model_selection import train_test_split     # Train/test split
from sklearn.linear_model import (                       # Linear models
    LogisticRegression, Lasso, Ridge, ElasticNet
)
from sklearn.svm import SVC                              # Support Vector Machine (SVM)
from sklearn.tree import DecisionTreeClassifier          # Decision Tree Classifier
from sklearn.naive_bayes import GaussianNB               # Gaussian Naive Bayes
from sklearn.neighbors import KNeighborsClassifier       # K-Nearest Neighbors (KNN)
from sklearn.ensemble import (                           # Ensemble models
    RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
)
from sklearn.metrics import (                            # Metrics for model evaluation
    confusion_matrix, f1_score, precision_score, recall_score, 
    accuracy_score, roc_auc_score, classification_report, 
    roc_curve, auc, r2_score, mean_absolute_error, mean_squared_error
)
```

```python
def evaluate_model(true, predicted):
    """
    Evaluate a regression model using different metrics.

    Parameters:
    - true: array-like, true target values
    - predicted: array-like, predicted target values

    Returns:
    - mae: float, mean absolute error
    - rmse: float, root mean squared error
    - r2_square: float, R-squared score
    """
    # Calculate mean absolute error
    mae = mean_absolute_error(true, predicted)
    
    # Calculate mean squared error
    mse = mean_squared_error(true, predicted)
    
    # Calculate root mean squared error
    rmse = np.sqrt(mse)
    
    # Calculate R-squared score
    r2_square = r2_score(true, predicted)
    
    return mae, rmse, r2_square
```

```python
# Define a dictionary of classification models to train
models = {
    'SVM': SVC(C=0.01, degree=5, kernel='poly'),                   # Support Vector Machine with polynomial kernel
    'KNN': KNeighborsClassifier(),                                 # K-Nearest Neighbors
    'DecisionTree': DecisionTreeClassifier(),                       # Decision Tree Classifier
    'GradientBoosting': GradientBoostingClassifier(),               # Gradient Boosting Classifier
    'LogisticRegression': LogisticRegression(max_iter=1000),        # Logistic Regression
    'AdaBoosting': AdaBoostClassifier(),                            # AdaBoost Classifier
    'Naive Bayes': GaussianNB()                                     # Gaussian Naive Bayes
}
```
